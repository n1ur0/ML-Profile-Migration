## b) inter-weeks distance to centroids difference - week 1 & week 2

We will now replace the distance to the centroids of each week, by the inter-weeks difference. 
We call this new feature **diff measure**.

I.e.:

**Time:**
- diff1_t = d1_y - d1_x (x represents week 1 & y represents week 2)
- diff2_t = d2_y - d2_x 
- diff3_t = d3_y - d3_x 
- diff4_t = d4_y - d4_x 
- diff5_t = d5_y - d5_x 
- diff6_t = d6_y - d6_x 

**Category:**
- diff1_c = d_cat1_y - d_cat1_x
- diff2_c = d_cat2_y - d_cat2_x
- diff3_c = d_cat3_y - d_cat3_x
- diff4_c = d_cat4_y - d_cat4_x
- diff5_c = d_cat5_y - d_cat5_x
- diff6_c = d_cat6_y - d_cat6_x

```python
# add "diff" columns
df_merge['diff1_t'] = df_merge['d1_y'] - df_merge['d1_x']
df_merge['diff2_t'] = df_merge['d2_y'] - df_merge['d2_x']
df_merge['diff3_t'] = df_merge['d3_y'] - df_merge['d3_x']
df_merge['diff4_t'] = df_merge['d4_y'] - df_merge['d4_x']
df_merge['diff5_t'] = df_merge['d5_y'] - df_merge['d5_x']
df_merge['diff6_t'] = df_merge['d6_y'] - df_merge['d6_x']

df_merge['diff1_c'] = df_merge['d_cat1_y'] - df_merge['d_cat1_x']
df_merge['diff2_c'] = df_merge['d_cat2_y'] - df_merge['d_cat2_x']
df_merge['diff3_c'] = df_merge['d_cat3_y'] - df_merge['d_cat3_x']
df_merge['diff4_c'] = df_merge['d_cat4_y'] - df_merge['d_cat4_x']
df_merge['diff5_c'] = df_merge['d_cat5_y'] - df_merge['d_cat5_x']
df_merge['diff6_c'] = df_merge['d_cat6_y'] - df_merge['d_cat6_x']
```

```python
# remove distances to centroid columns
df_merge.drop(['d1_x', 'd2_x', 'd3_x', 'd4_x', 'd5_x', 'd6_x',
               'd1_y', 'd2_y', 'd3_y', 'd4_y', 'd5_y', 'd6_y',
               'd_cat1_x', 'd_cat2_x', 'd_cat3_x', 'd_cat4_x', 'd_cat5_x', 'd_cat6_x',
               'd_cat1_y', 'd_cat2_y', 'd_cat3_y', 'd_cat4_y', 'd_cat5_y', 'd_cat6_y'], 
              axis=1, inplace=True)
```

## features slot

This section implements a custom function that: 

**a)** filters devices with activity in the two weeks (current & next week) & 

**b)** tracks devices' migration (do they change cluster?) 

**c)** computes the inter-weeks difference regarding the distance to centroids  

```python
# auxiliary function to add binary labels
def binary_label(x):
    if x == 0: 
        return 0 # device does not change cluster
    else:
        return 1 # device does change cluster
```

```python
def features(df_currentWeek, df_nextWeek):
    """
    This function:
    a) filters devices with activity in the two weeks (current & next week) 
    b) tracks devices' migration (do they change cluster? 0/1)
    c) replaces the distances to the centroids by the inter-weeks difference
    """
    # merge weeks based on devices that have activity on both periods (inner join)
    df_merge = pd.merge(df_currentWeek, df_nextWeek, how='inner', left_on='device', right_on='device')
    
    # add "change" columns to df_features
    df_merge['change_TimeCluster'] = df_merge['TimeCluster_x'] - df_merge['TimeCluster_y']
    df_merge['change_TimeCluster'] = df_merge['change_TimeCluster'].apply(binary_label)

    df_merge['change_CatCluster'] = df_merge['CatCluster_x'] - df_merge['CatCluster_y']
    df_merge['change_CatCluster'] = df_merge['change_CatCluster'].apply(binary_label)
    
    # add "diff" columns
    df_merge['diff1_t'] = df_merge['d1_y'] - df_merge['d1_x']
    df_merge['diff2_t'] = df_merge['d2_y'] - df_merge['d2_x']
    df_merge['diff3_t'] = df_merge['d3_y'] - df_merge['d3_x']
    df_merge['diff4_t'] = df_merge['d4_y'] - df_merge['d4_x']
    df_merge['diff5_t'] = df_merge['d5_y'] - df_merge['d5_x']
    df_merge['diff6_t'] = df_merge['d6_y'] - df_merge['d6_x']

    df_merge['diff1_c'] = df_merge['d_cat1_y'] - df_merge['d_cat1_x']
    df_merge['diff2_c'] = df_merge['d_cat2_y'] - df_merge['d_cat2_x']
    df_merge['diff3_c'] = df_merge['d_cat3_y'] - df_merge['d_cat3_x']
    df_merge['diff4_c'] = df_merge['d_cat4_y'] - df_merge['d_cat4_x']
    df_merge['diff5_c'] = df_merge['d_cat5_y'] - df_merge['d_cat5_x']
    df_merge['diff6_c'] = df_merge['d_cat6_y'] - df_merge['d_cat6_x']
    
    # remove distances to centroid columns
    #df_merge.drop(['d1_x', 'd2_x', 'd3_x', 'd4_x', 'd5_x', 'd6_x',
                   #'d1_y', 'd2_y', 'd3_y', 'd4_y', 'd5_y', 'd6_y',
                   #'d_cat1_x', 'd_cat2_x', 'd_cat3_x', 'd_cat4_x', 'd_cat5_x', 'd_cat6_x',
                   #'d_cat1_y', 'd_cat2_y', 'd_cat3_y', 'd_cat4_y', 'd_cat5_y', 'd_cat6_y'], 
                  #axis=1, inplace=True)
    
    # remove labels
    df_merge.drop(['TimeCluster_x','TimeCluster_y','CatCluster_x','CatCluster_y'],
                  axis=1, inplace=True)
    
    return df_merge  
```

# Labels slot

In this section, we prepare the labels' slot.

```python
# STEP 1 
df_features_devices = slot_features['device'] # devices with activity in the features'weeks
df_labels = pd.merge(df_features_devices, df_w3, how='left', left_on='device', right_on='device')

# sanity check
print(df_labels.shape)
print(df_labels.columns)
```

```python
# STEP 2
# keep only the columns: 'device', 'TimeCluster', 'CatCluster'
df_labels = df_labels[['device', 'TimeCluster', 'CatCluster']]
df_labels
```

We also add 2 new labels: **"Change_TimeCluster"** & **"Change_CatCluster"**.

Each one has 3 classes that, regarding the week previous to the week label, identify:

- **"0"**, devices that don't change cluster
- **"1"**, devices that change cluster
- **"missing"**, devices that do not have activity

```python
# auxiliary function to add binary labels
def binary_label(x):
    if x == 0: 
        return 0 # device does not change cluster
    elif (x>0 or x<0):
        return 1  # device does change cluster
    else:
        return "missing" 
```

```python
# STEP 3
# new labels: "Change_TimeCluster" & "Change_CatCluster" 
df_labels = pd.merge(df_labels, df_w2, how='left', left_on='device', right_on='device')

# add "change" columns to df_labels
df_labels['Change_TimeCluster'] = df_labels['TimeCluster_x'] - df_labels['TimeCluster_y'] #**
df_labels['Change_TimeCluster'] = df_labels['Change_TimeCluster'].apply(binary_label) # add binary label
#** the device does not change cluster only if this difference is equal to 0

df_labels['Change_CatCluster'] = df_labels['CatCluster_x'] - df_labels['CatCluster_y']
df_labels['Change_CatCluster'] = df_labels['Change_CatCluster'].apply(binary_label)

df_labels
```

```python
df_labels.columns
```

```python
# STEP 4
# add the label "missing" 

df_labels['TimeCluster_x'].fillna('missing', inplace=True)
df_labels['CatCluster_x'].fillna('missing', inplace=True)

df_labels['Change_TimeCluster'].fillna('missing', inplace=True)
df_labels['Change_CatCluster'].fillna('missing', inplace=True)

# keep only the columns "device", 'TimeCluster_x', 'CatCluster_x', 
# 'Change_TimeCluster', 'Change_CatCluster'

df_labels = df_labels[['device', 'TimeCluster_x', 'CatCluster_x', 'Change_TimeCluster', 'Change_CatCluster' ]]
df_labels = df_labels.rename(columns = {'TimeCluster_x': 'TimeCluster_lab', 'CatCluster_x': 'CatCluster_lab','Change_TimeCluster': 'Change_TimeCluster_lab','Change_CatCluster': 'Change_CatCluster_lab'}, inplace = False)

df_labels
```

Gathering the previous steps in a custom function:

```python
def labels(slot_features, df_previous_week, df_week_label):
    """
    This function prepares the labels slot.
    """
    # STEP 1 
    df_features_devices = slot_features['device'] # devices with activity in the features'weeks
    df_labels = pd.merge(df_features_devices, df_week_label, how='left', left_on='device', right_on='device')
    
    # STEP 2
    df_labels = df_labels[['device', 'TimeCluster', 'CatCluster']]
    
    # STEP 3
    # new labels: "Change_TimeCluster" & "Change_CatCluster" 
    df_labels = pd.merge(df_labels, df_previous_week, how='left', left_on='device', right_on='device')

    # add "change" columns to df_labels
    df_labels['Change_TimeCluster'] = df_labels['TimeCluster_x'] - df_labels['TimeCluster_y'] #**
    df_labels['Change_TimeCluster'] = df_labels['Change_TimeCluster'].apply(binary_label) # add binary label
    #** the device does not change cluster only if this difference is equal to 0

    df_labels['Change_CatCluster'] = df_labels['CatCluster_x'] - df_labels['CatCluster_y']
    df_labels['Change_CatCluster'] = df_labels['Change_CatCluster'].apply(binary_label)
    
    # STEP 4
    # add the label "missing" 

    df_labels['TimeCluster_x'].fillna('missing', inplace=True)
    df_labels['CatCluster_x'].fillna('missing', inplace=True)

    df_labels['Change_TimeCluster'].fillna('missing', inplace=True)
    df_labels['Change_CatCluster'].fillna('missing', inplace=True)

    # keep only the columns "device", 'TimeCluster_x', 'CatCluster_x', 
    # 'Change_TimeCluster', 'Change_CatCluster'

    df_labels = df_labels[['device', 'TimeCluster_x', 'CatCluster_x', 'Change_TimeCluster', 'Change_CatCluster' ]]
    df_labels = df_labels.rename(columns = {'TimeCluster_x': 'TimeCluster_lab', 'CatCluster_x': 'CatCluster_lab',
                                            'Change_TimeCluster': 'Change_TimeCluster_lab',
                                            'Change_CatCluster': 'Change_CatCluster_lab',}, inplace = False)
    
    return df_labels
       
```

```python
# example slot_labels for week 3
slot_labels = labels(slot_features, df_w2, df_w3)

slot_labels
```
